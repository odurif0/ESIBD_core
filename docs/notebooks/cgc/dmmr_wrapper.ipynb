# CGC DMMR Notebook

Documentation and manual test notebook for the CGC `DMMR` wrapper.

Structure:
- one simple end-to-end sequence
- one connected-session section for read-only and configuration-inspection calls
- one measurement-session section for module current acquisition and housekeeping helpers
- one low-level transport section for `open_port()` / `close_port()` style calls


```python
import time

from cgc.dmmr import DMMR

DEVICE_ID = "DMMR_TEST"
COM_PORT = 8
BAUDRATE = 230400
HK_INTERVAL_S = 3.0
TIMEOUT_S = 5.0

MODULE_ADDRESS = 0
CONFIG_NUMBER = 0
TEST_MEAS_RANGE = 0
```

## 1. Simple Sequence

Recommended high-level sequence: instantiate, connect, scan modules, inspect one module,
toggle automatic current acquisition, then shutdown.


```python
dmmr_simple = DMMR(device_id=DEVICE_ID, com=COM_PORT)  # Optional arguments: baudrate=BAUDRATE, hk_interval_s=HK_INTERVAL_S
dmmr_simple.connect()  # Optional argument: timeout_s=TIMEOUT_S
modules = dmmr_simple.scan_modules()
print(f"Detected modules: {modules}")
```

```python
module_address = next(iter(modules), MODULE_ADDRESS)
print("dmmr_simple.get_module_info(module_address)", dmmr_simple.get_module_info(module_address))
print("dmmr_simple.set_enable(True)", dmmr_simple.set_enable(True))
print("dmmr_simple.set_automatic_current(True)", dmmr_simple.set_automatic_current(True))
time.sleep(0.5)
print("dmmr_simple.get_current()", dmmr_simple.get_current())  # NO_DATA is normal when no fresh frame is available yet.
```

```python
dmmr_simple.shutdown()  # disables acquisition and disconnects from the instrument.
dmmr_simple.close()  # releases Python-side worker/process resources.
```
