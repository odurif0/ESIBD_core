# CGC AMX Notebook

Documentation and manual test notebook for the CGC `AMX` wrapper.

Structure:
- one simple end-to-end sequence
- one connected-session section for wrapper methods
- one configured-session section for timing and pulser helpers
- one low-level transport section for `open_port()` / `close_port()` style calls


```python
from pathlib import Path
import sys

current = Path.cwd().resolve()
for candidate in (current, *current.parents):
    if (candidate / "src").exists():
        REPO_ROOT = candidate
        break
else:
    raise RuntimeError("Could not locate the repository root containing src/.")

SRC_DIR = REPO_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

print(f"Repo root: {REPO_ROOT}")
print(f"Using source tree: {SRC_DIR}")
```

```python
from cgc.amx import AMX

DEVICE_ID = "AMX_MAIN"
COM_PORT = 8
PORT_NUMBER = 0
BAUDRATE = 230400
CONFIG_NUMBER = 40
TIMEOUT_S = 5.0

PULSER_NO = 0
SWITCH_NO = 0
FREQUENCY_HZ = 2000.0
FREQUENCY_KHZ = 2.0
DUTY_CYCLE = 0.5
PULSER_DELAY_TICKS = 100
PULSER_WIDTH_TICKS = 100
SWITCH_ENABLE_DELAY_TICKS = 10
SWITCH_TRIGGER_RISE_TICKS = 10
SWITCH_TRIGGER_FALL_TICKS = 20


def show(label, value):
    print(label)
    print(value)
    print()
```

## 1. Simple Sequence

Recommended sequence: instantiate, connect, load a known config,
tweak a pulser setting, then shutdown.


```python
amx_simple = AMX(
    device_id=DEVICE_ID,
    com=COM_PORT,
    port=PORT_NUMBER,
    baudrate=BAUDRATE,
)

try:
    amx_simple.connect(timeout_s=TIMEOUT_S)
    amx_simple.load_config(CONFIG_NUMBER)
    amx_simple.set_frequency_hz(FREQUENCY_HZ)
    amx_simple.set_pulser_duty_cycle(PULSER_NO, DUTY_CYCLE)
    show("amx_simple.get_frequency_hz()", amx_simple.get_frequency_hz())
finally:
    amx_simple.shutdown()
    amx_simple.close()
```
