# CGC PSU Notebook

Documentation and manual test notebook for the CGC `PSU` wrapper.

Structure:
- one simple end-to-end sequence
- one connected-session section for wrapper methods
- one configured-session section for channel control helpers
- one low-level transport section for `open_port()` / `close_port()` style calls


```python
from pathlib import Path
import sys

current = Path.cwd().resolve()
for candidate in (current, *current.parents):
    if (candidate / "src").exists():
        REPO_ROOT = candidate
        break
else:
    raise RuntimeError("Could not locate the repository root containing src/.")

SRC_DIR = REPO_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

print(f"Repo root: {REPO_ROOT}")
print(f"Using source tree: {SRC_DIR}")
```

```python
from cgc.psu import PSU

DEVICE_ID = "PSU_MAIN"
COM_PORT = 6
PORT_NUMBER = 0
BAUDRATE = 230400
CONFIG_NUMBER = 19
TIMEOUT_S = 5.0

CHANNEL = 0
TEST_VOLTAGE_V = 5.0
TEST_CURRENT_A = 0.1


def show(label, value):
    print(label)
    print(value)
    print()
```

## 1. Simple Sequence


```python
psu_simple = PSU(
    device_id=DEVICE_ID,
    com=COM_PORT,
    port=PORT_NUMBER,
    baudrate=BAUDRATE,
)

try:
    psu_simple.connect(timeout_s=TIMEOUT_S)
    psu_simple.load_config(CONFIG_NUMBER)
    psu_simple.set_channel_voltage(CHANNEL, TEST_VOLTAGE_V)
    psu_simple.set_channel_current(CHANNEL, TEST_CURRENT_A)
    show("psu_simple.get_channel_voltage(CHANNEL)", psu_simple.get_channel_voltage(CHANNEL))
    show("psu_simple.get_channel_current(CHANNEL)", psu_simple.get_channel_current(CHANNEL))
finally:
    psu_simple.shutdown()
    psu_simple.close()
```
